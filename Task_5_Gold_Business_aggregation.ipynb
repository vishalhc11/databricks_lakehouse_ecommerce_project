**1. Sales by state or region**

```python
%sql
select customer_state,  upper(customer_city) customer_city , sum(revenue) as state_wise_sales_revenue from fact_sales group by customer_state , customer_city order by state_wise_sales_revenue desc ;
```

```python
df_sales_by_state_city = spark.sql("select customer_state,  upper(customer_city) customer_city , sum(revenue) as state_wise_sales_revenue from fact_sales group by customer_state , customer_city order by state_wise_sales_revenue desc")
```

```python
df_sales_by_state_city.limit(10).display()
```

```python
df_sales_by_state_city.write.mode("overwrite").parquet("/Volumes/main/ecommerce/lakehouse_volumes/gold_dataset/sales_by_state_city/")
```

**2. Top products by revenue**

```python
%sql
select product_id ,  product_category_name , sum(revenue) as top_products_by_revenue from fact_sales group by product_id ,  product_category_name order by top_products_by_revenue desc ;
```

```python
df_top_products_by_revenue = spark.sql("select product_id ,  product_category_name , sum(revenue) as top_products_by_revenue from fact_sales group by product_id ,  product_category_name order by top_products_by_revenue desc") ;
```

```python
df_top_products_by_revenue.write.mode("overwrite").parquet("/Volumes/main/ecommerce/lakehouse_volumes/gold_dataset/top_products_by_revenue/")
```

**3. Daily or monthly sales trends**

```python
%sql
desc fact_sales
```

```python
%sql
-- note date_format to convert date into string 

select  date_format(order_date, 'yyyy-MM')  order_months , 
        product_category_name, 
        sum(revenue) sum_of_sales 
from  fact_sales
group by date_format(order_date, 'yyyy-MM'), 
         product_category_name order by 1 , 3 desc
```

```python
df_monthly_sales_trends = spark.sql("""select date_format(order_date, 'yyyy-MM')  order_months , 
        product_category_name, 
        sum(revenue) sum_of_sales 
from  fact_sales
group by date_format(order_date, 'yyyy-MM'), 
         product_category_name order by 1 , 3 desc""")
```

```python
df_monthly_sales_trends.display()
```

```python
df_monthly_sales_trends.write.mode("overwrite").parquet("/Volumes/main/ecommerce/lakehouse_volumes/gold_dataset/monthly_sales_trends/")
```
